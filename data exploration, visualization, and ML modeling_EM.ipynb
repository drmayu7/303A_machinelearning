# ML Models: based on splitting data into 80% train set and 20% test set

```python
from sklearn.preprocessing import MinMaxScaler

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn import metrics # for the check the error and accuracy

from sklearn import svm # for Support Vector Machine
from sklearn.naive_bayes import GaussianNB # for naive bayes classifier
from sklearn.neighbors import KNeighborsClassifier # for KNN classifier 
from sklearn.ensemble import RandomForestClassifier # for random forest classifier
from sklearn.tree import DecisionTreeClassifier # for decision tree classifier 
```

```python
# Split the dataset into features/variables (X) and labels/outcome (y)
# Change values of 'class' from [2,4] to [0,1]
X = df.drop(columns=['class'])
y = df['class'].apply(lambda x: 1 if x == 4 else 0)

print('X = ', X)
print('Y = ', y)
```

```python
# Scale the features - Min and Max scaling (sensitive to outliers)
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

# Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

#split data into train set and test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state = 0)

# define the model
model = LogisticRegression()

# fit on the training set
model.fit(X_train, y_train)

# predict on test set
pred_1 = model.predict(X_test)

# evaluate predictions
accuracy_1 = accuracy_score(y_test, pred_1)

print('Accuracy: %.2f' % (accuracy_1*100))
```

# Support Vector Machine 

```python
from sklearn import svm

#split data into train set and test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state = 0)

# define the model
clf=svm.SVC()

# fit on the training set
clf.fit(X_train,y_train)

# predict on test set
pred_2 = clf.predict(X_test)

# evaluate predictions
accuracy_2 = accuracy_score(y_test, pred_2)

print('Accuracy: %.2f' % (accuracy_2*100))
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

#split data into train set and test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state = 0)

# define the model
nb = GaussianNB()

# fit on the training set
nb.fit(X_train,y_train)

# predict on test set
pred_3 = nb.predict(X_test)

# evaluate predictions
accuracy_3 = accuracy_score(y_test, pred_3)

print('Accuracy: %.2f' % (accuracy_3*100))
    
```

# KNN Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

#split data into train set and test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state = 0)

# define the model
knn=KNeighborsClassifier(n_neighbors=5)

# fit on the training set
knn.fit(X_train, y_train)

# predict on test set
pred_4 = knn.predict(X_test)

# evaluate predictions
accuracy_4 = accuracy_score(y_test, pred_4)

print('Accuracy: %.2f' % (accuracy_4*100))
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

#split data into train set and test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state = 0)

# define the model
rfc=RandomForestClassifier()

# fit on the training set
rfc.fit(X_train,y_train)

# predict on test set
pred_5 = rfc.predict(X_test)

# evaluate predictions
accuracy_5 = accuracy_score(y_test, pred_5)

print('Accuracy: %.2f' % (accuracy_3*100))
```
