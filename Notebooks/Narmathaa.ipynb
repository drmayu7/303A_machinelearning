# Descriptive Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import os 
os.getcwd()
```

```python
#---Reading the dataset using URL
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data"
df = pd.read_csv(url,index_col=None,header=None)
df #---The breast cancer wisconsin original dataset
```

```python
#---Renaming the column names using the attributes from breast-cancer-wisconsin.names
#---Create a function

def rename(df):
    header_dict = {
        0:'sample_id',
        1:'cl_thcknss',
        2:'size_cell_un',
        3:'shape_cell_un',
        4:'marg_adhesion',
        5:'size_cell_single',
        6:'bare_nucl',
        7:'bl_chrmatn',
        8:'nrml_nucleo',
        9:'mitoses',
        10:'class'
    }
    df.rename(columns=header_dict,inplace=True)
    
    #replace ? values with NaN
    df.replace({'?':'NaN'},inplace=True)
    
rename(df)
print('Breast cancer Wisconsin dataset with columns renamed according to the attributes and "?" replaced with "NaN":')
df
```

```python
#---To view the top 5 rows of the dataset
#---To confirm the values are same as the original dataset
print('The top 5 rows of the dataset:')
df.head()
```

```python
#---To view the bottom 5 rows of the dataset
#---To confirm the values are same as the original dataset
print('The bottom 5 rows of the dataset:')
df.tail()
```

```python
#---Remove the sample_id column
df_2 = df.drop(columns=['sample_id'])
print('Breast cancer Wisconsin dataset without sample_id column: ')
df_2 #---Breast cancer wisconsin dataset WITHOUT the sample_id column
```

```python
df_2.describe()
```

```python
#---Pearson's Test using seaborn
corr = df_2.corr()
f, ax = plt.subplots(figsize=(8, 5))
mask = np.triu(np.ones_like(corr, dtype=bool))
cmap = sb.diverging_palette(230, 20, as_cmap=True)
sb.heatmap(corr, annot=True, mask = mask, cmap=cmap)
plt.title("Correlation Coefficient (Pearson-Test) for Multivariate Features of Breast Cancer Wisconsin Dataset")
plt.show()
```

```python
#---Visualising a correlation heatmap using seaborn
print("A correlation heatmap is shown as below:")
plt.figure(figsize=(8,6))
sb.set_theme(style="white")
corr = df_2.corr()
heatmap = sb.heatmap(corr, annot=True, cmap="BuPu", fmt='.1g')
plt.title("Heatmap of Breast cancer Wisconsin Dataset")
plt.show()
```
