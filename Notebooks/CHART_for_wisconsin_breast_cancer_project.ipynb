```python
# importing library
import matplotlib.pyplot as plt
  
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
  
if __name__ == '__main__':
    
    # creating data on which bar chart will be plot
    x = ["RF", "NB", "DT", "Logistic Reg", "kNN", "SVM", "Neural Network"]
    y = [0.97,0.96,0.95,0.96,0.96,0.96,0.97]
      
    # setting figure size by using figure() function 
    plt.figure(figsize = (10, 5))
      
    # making the bar chart on the data
    plt.bar(x, y)
      
    # calling the function to add value labels
    addlabels(x, y)
      
    # giving title to the plot
    plt.title("Comparison of accuracy for all Classifiers")
      
    # giving X and Y labels
    plt.xlabel("ML Classifier")
    plt.ylabel("accuracy score")
      
    # visualizing the plot
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import lines
from matplotlib import patches
from matplotlib.patheffects import withStroke

accuracy_score = [97,96,95,96,96,96,97]
classifier= ["RF", "NB", "DT", "Logistic Reg", "kNN", "SVM", "Neural Network"]

# The positions for the bars
# This allows us to determine exactly where each bar is located
y = [i * 0.9 for i in range(len(classifier))]

# The colors
BLUE = "#076fa2"
RED = "#E3120B"
BLACK = "#202020"
GREY = "#a2a2a2"

fig, ax = plt.subplots(figsize=(12, 7))

ax.barh(y, accuracy_score, height=0.55, align="edge", color=BLUE);

ax.xaxis.set_ticks([i * 10 for i in range(0, 12)])
ax.xaxis.set_ticklabels([i * 10 for i in range(0, 12)], size=16, fontfamily="Econ Sans Cnd", fontweight=100)
ax.xaxis.set_tick_params(labelbottom=False, labeltop=True, length=0)

ax.set_xlim((0, 120))
ax.set_ylim((0, len(classifier) * 0.9 - 0.2))

# Set whether axis ticks and gridlines are above or below most artists.
ax.set_axisbelow(True)
ax.grid(axis = "x", color="#A8BAC4", lw=1.2)
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["left"].set_lw(1.5)
# This capstyle determines the lines don't go beyond the limit we specified
# see: https://matplotlib.org/stable/api/_enums_api.html?highlight=capstyle#matplotlib._enums.CapStyle
ax.spines["left"].set_capstyle("butt")

# Hide y labels
ax.yaxis.set_visible(False)

PAD = 0.3
for name, count, y_pos in zip(classifier, accuracy_score, y):
    x = 0
    color = "white"
    path_effects = None
    if count < 8:
        x = count
        color = BLUE    
        path_effects=[withStroke(linewidth=6, foreground="white")]
    
    ax.text(
        x + PAD, y_pos + 0.5 / 2, name, 
        color=color, fontfamily="Econ Sans Cnd", fontsize=18, va="center",
        path_effects=path_effects
    ) 

# Make room on top and bottom
# Note there's no room on the left and right sides
fig.subplots_adjust(left=0.005, right=1, top=0.8, bottom=0.1)

# Add title
fig.text(
    0, 0.925, "Accuracy of Classifier", 
    fontsize=22, fontweight="bold", fontfamily="Econ Sans Cnd"
)
# Add subtitle
fig.text(
    0, 0.875, "Accuracy of Classifier in wisconsin breast cancer dataset", 
    fontsize=20, fontfamily="Econ Sans Cnd"
)


# Add line and rectangle on top.
fig.add_artist(lines.Line2D([0, 1], [1, 1], lw=3, color=RED, solid_capstyle="butt"))
fig.add_artist(patches.Rectangle((0, 0.975), 0.05, 0.025, color=RED))

# Set facecolor, useful when saving as .png
fig.set_facecolor("white")


#fig.savefig("plot.png", dpi=300)
```

The comparison of Sensitivity Performance among ML Classifier

```python
import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['RF','NB','DT','Logistic Reg','kNN','SVM']

benign = [97.16,97.32,96.29,97.60,97.16,94.10]
malignant= [97.16,95.19,96.29,97.60,97.16,94.10]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, benign, 0.4, label = 'B')
plt.bar(X_axis + 0.2, malignant, 0.4, label = 'M')
  
plt.xticks(X_axis, X)
plt.xlabel("ML Classifier")
plt.ylabel("Sensitivity")
plt.title("The comparison of Sensitivity Performance among ML Classifier")
plt.legend()
plt.show()
```

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
classifier = ['RF','NB','DT','Logistic Reg','kNN','SVM']
benign = [97.16,97.32,96.29,97.60,97.16,94.10]
malignant = [97.16,95.19,96.29,97.60,97.16,94.10]
# Data
df=pd.DataFrame({'classifier': range(1,11), 'benign': np.random.randn(10), 'malignant': np.random.randn(10)+range(1,11)})
 
# multiple line plots
plt.plot( 'classifier', 'benign', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'classifier', 'malignant', data=df, marker='', color='olive', linewidth=2)
# Add titles
plt.title("The comparison of Sensitivity Performance among ML Classifier", loc='left', fontsize=12, fontweight=0, color='black')
plt.xlabel("classifier")
plt.ylabel("Sensitivity")

# show legend
plt.legend()

# show graph
plt.show()
```
