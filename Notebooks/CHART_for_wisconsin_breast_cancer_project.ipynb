ACCURACY

```python
# importing library
import matplotlib.pyplot as plt
  
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
  
if __name__ == '__main__':
    
    # creating data on which bar chart will be plot
    x = ["RF", "NB", "DT", "Logistic Reg", "kNN", "SVM", "Neural Network"]
    y = [0.97,0.96,0.95,0.96,0.96,0.96,0.97]
      
    # setting figure size by using figure() function 
    plt.figure(figsize = (10, 5))
      
    # making the bar chart on the data
    plt.bar(x, y)
      
    # calling the function to add value labels
    addlabels(x, y)
      
    # giving title to the plot
    plt.title("Comparison of accuracy for all Classifiers")
      
    # giving X and Y labels
    plt.xlabel("ML Classifier")
    plt.ylabel("accuracy score")
      
    # visualizing the plot
    plt.show()
```

TRUE POSITIVE & FALSE POSITIVE

```python
import numpy as np 
import matplotlib.pyplot as plt 
  
X = ['RF','NB','DT','Logistic Reg','kNN','SVM']

TRUE_POSITIVE = [445,436,441,447,445,431]
FALSE_POSITIVE= [7,7,21,14,12,3]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, TRUE_POSITIVE, 0.4, label = 'TP')
plt.bar(X_axis + 0.2, FALSE_POSITIVE, 0.4, label = 'FP')
  
plt.xticks(X_axis, X)
plt.xlabel("ML Classifier")
plt.ylabel("Number of samples")
plt.title("The comparison of True Positive & False Positive Among Classifiers")
plt.legend()
plt.show()
```
